Q1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?

-> Basic Filters use simple selections like checkboxes or dropdowns in the Filters pane for quick field-based limiting. Advanced Filters provide conditional options such as "Top N," "greater than," or text contains for more nuanced control. Slicers are interactive visual elements (buttons/lists) placed on the canvas, enabling users to filter multiple visuals dynamically across pages.

Q4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?

-> Visual-Level Filters apply only to a single visual, refining its data independently. Page-Level Filters affect all visuals on one specific page, while Report-Level Filters impact every visual across the entire report (e.g., limiting to India sales).

Q6. What is the role of Drill-Down in Power BI? How does it improve data exploration?

-> Drill-Down enables hierarchical navigation within a single visual, expanding from summary (e.g., Country) to details (e.g., Product Category) by clicking icons or data points. It improves exploration by allowing seamless, interactive drilling into data layers without switching visuals or pages, revealing trends and outliers progressively.

Q10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.

-> Drill-Down expands hierarchy levels within the same visual (e.g., Country → Region → Product Category) by clicking drill icons. Drill-Through navigates to a separate detail page filtered by the selected item (right-click context menu).

Real-World Example: In sales analysis, Drill-Down on a national sales map reveals state-level data instantly. Drill-Through from a specific state bar takes you to a dedicated page with that state's customer breakdowns, KPIs, and time trends.

Q13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction?

-> The "Keep All Filters" option in Drill-Through preserves all slicers, page filters, and visual filters from the source page when navigating to the target detail page. This maintains full context (e.g., selected Year or Region carries over), preventing data loss and enabling consistent analysis across pages.

User Interaction Impact: Users see filtered drill-through pages that respect their current selections, creating seamless workflows without manual re-filtering—ideal for focused deep dives within existing constraints.
